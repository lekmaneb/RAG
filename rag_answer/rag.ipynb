```python
# open pdf file and extract text
import PyPDF2
import sys

def pdf_text_extractor(pdf_file):
    pdf_file = open(pdf_file, 'rb')
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    pdf_text = ''
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        pdf_text += page.extract_text()
    pdf_file.close()
    return pdf_text
```

```python
# open wikipedia page and extract text
import requests
from bs4 import BeautifulSoup

def wiki_text_extractor(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    wiki_text = ''
    for paragraph in soup.find_all('p'):
        wiki_text += paragraph.get_text()
    return wiki_text
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Define the Hugging Face model name
# hf_model_name = "Qwen/Qwen2.5-3B-Instruct"  # Replace with a larger or more conversational model as needed
hf_model_name = "Qwen/Qwen2.5-1.5B-Instruct"
# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(hf_model_name)
model = AutoModelForCausalLM.from_pretrained(hf_model_name).to("cuda")  # Move model to GPU

# Create a pipeline for text generation with GPU support
hf_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128,
    truncation=True,
    device=0  # Use GPU (device index 0)
)

# Wrap the pipeline in a HuggingFacePipeline
llm = HuggingFacePipeline(pipeline=hf_pipeline)
```

```python
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate

# Build prompt
template = """Use the following pieces of context to answer the question at the end. If you don't know the answer, just say that you don't know, don't try to make up an answer. Use three sentences maximum. Keep the answer as concise as possible. Always say "thanks for asking!" at the end of the answer. 
{context}
Question: {question}
Helpful Answer:"""
QA_CHAIN_PROMPT = PromptTemplate.from_template(template)# Run chain

qa_chain = RetrievalQA.from_chain_type(
    llm,
    return_source_documents=True,
    retriever=vectordb.as_retriever(search_kwargs={"k": 1}),
    chain_type_kwargs={"prompt": QA_CHAIN_PROMPT}
)
```

```python
# Pass question to the qa_chain
question = "What's Pansy's age?"
result = qa_chain({"query": question})
result["result"]
```

```python
import pandas as pd
import os

# music data folder contains csv files. open all csv, take the 20 first rows, keep only the columns 'Artist', 'Title' and 'Lyrics' and concatenate them in a single dataframe
music_data_folder = "music_data"
dfs = []
for file in os.listdir(music_data_folder):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(music_data_folder, file)) #, nrows=20)
        df = df[['Artist', 'Title', 'Lyric']]
        dfs.append(df)

music_df = pd.concat(dfs, ignore_index=True)

music_df.describe()
```

```python
# save the dataframe in a csv file
music_df.to_csv("all_music_data.csv", index=False)
```

```python
# now transform the dataframe into a list of list containing ['Name of the song: {Title}', 'Artist: {Artist}', 'Lyrics: {Lyric}']
music_data = []
for idx, row in music_df.iterrows():
    music_data.append(f"Name of the artist: {row['Artist']}; Name of the song: {row['Title']}; Lyrics of the song: {row['Lyric']}")
```
